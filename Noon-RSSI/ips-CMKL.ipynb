### Separate Models Approach

```python
from sklearn.multioutput import MultiOutputRegressor

# Define models for comparison
models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(),
    "Random Forest": RandomForestRegressor(),
    "Gradient Boosting": GradientBoostingRegressor(),
    "K-Neighbors": KNeighborsRegressor()
}

# K-Fold for validation
kf = KFold(n_splits=5, random_state=42, shuffle=True)

# Train separate models for Latitude and Longitude and predict
predictions = {}
for name, model in models.items():
    predictions[name] = {}
    for target in ['Latitude', 'Longitude']:
        # Fit the model on the training data
        model.fit(X_train, y_train[target])
        # Predict on the test data
        predicted = model.predict(X_test)
        # Store predictions
        predictions[name][target] = predicted
        # Evaluate model using cross-validation on the training set
        cv_scores = cross_val_score(model, X_train, y_train[target], cv=kf, scoring='neg_mean_squared_error')
        print(f"{name} {target} - CV MSE: {-cv_scores.mean():.3f} +/- {cv_scores.std():.3f}")

# Display RSSI values, predictions, and actual values for comparison
for name, model_preds in predictions.items():
    print(f"Model: {name}")
    for idx, (features, actual) in enumerate(zip(X_test.values, y_test.values)):
        predicted_lat = model_preds['Latitude'][idx]
        predicted_lon = model_preds['Longitude'][idx]
        print(f"RSSI values: {features}")
        print(f"Predicted Latitude: {predicted_lat}, Actual Latitude: {actual[0]}")
        print(f"Predicted Longitude: {predicted_lon}, Actual Longitude: {actual[1]}")
        print('-' * 60)
```

#### Separate Models with randomized search

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from scipy.stats import randint as sp_randint
from scipy.stats import uniform

# Models setup
models = {
    'Latitude': {
        'Decision Tree': DecisionTreeRegressor(),
        'Random Forest': RandomForestRegressor(),
        'Linear Regression': LinearRegression(),
        'Gradient Boosting': GradientBoostingRegressor(),
        'K-Neighbors': KNeighborsRegressor()
    },
    'Longitude': {
        'Decision Tree': DecisionTreeRegressor(),
        'Random Forest': RandomForestRegressor(),
        'Linear Regression': LinearRegression(),
        'Gradient Boosting': GradientBoostingRegressor(),
        'K-Neighbors': KNeighborsRegressor()
    }
}

# Parameter distributions for Random Search
param_distributions = {
    'Decision Tree': {
        'max_depth': sp_randint(3, 20),
        'min_samples_split': sp_randint(2, 20)
    },
    'Random Forest': {
        'n_estimators': sp_randint(50, 200),
        'max_features': [None, 'sqrt', 'log2']
    },
    'Linear Regression': {
    },
    'Gradient Boosting': {
        'n_estimators': sp_randint(100, 300),
        'learning_rate': uniform(0.01, 0.19),  # Adjusted upper bound to keep within 0.2 max
        'subsample': uniform(0.8, 0.2)  # Corrected to ensure values <= 1.0
    },
    'K-Neighbors': {
        'n_neighbors': sp_randint(3, 10),
        'weights': ['uniform', 'distance']
    }
}

# Dictionary to store the best parameters
best_params = {'Latitude': {}, 'Longitude': {}}

# Perform Random Search for each model and output
for output, configs in models.items():
    for name, model in configs.items():
        random_search = RandomizedSearchCV(model, param_distributions[name], n_iter=50, cv=5, scoring='neg_mean_squared_error', random_state=42)
        random_search.fit(X_train, y_train[output])  # Assuming y_train[output] is the correct target column
        best_params[output][name] = random_search.best_params_
        print(f"Best parameters for {name} predicting {output}: {random_search.best_params_}")
        print(f"Best score for {name} predicting {output}: {random_search.best_score_}")

print(best_params)
```

### MultiOutputRegressor Approach

```python
from sklearn.multioutput import MultiOutputRegressor

# Define models for comparison using MultiOutputRegressor
models = {
    "Linear Regression": MultiOutputRegressor(LinearRegression()),
    "Decision Tree": MultiOutputRegressor(DecisionTreeRegressor()),
    "Random Forest": MultiOutputRegressor(RandomForestRegressor()),
    "Gradient Boosting": MultiOutputRegressor(GradientBoostingRegressor()),
    "K-Neighbors": MultiOutputRegressor(KNeighborsRegressor())
}

results = {}

# Evaluate each model using K-fold cross-validation
for name, model in models.items():
    model_results = cross_val_score(model, X_train, y_train, cv=kf, scoring='neg_mean_squared_error')
    results[name] = model_results
    print(f"{name} CV MSE: {-model_results.mean():.3f} +/- {model_results.std():.3f}")

for name, model in models.items():
    model.fit(X_train, y_train)  # Fit model on training data
    predictions = model.predict(X_test)  # Predict on test data
    # Display RSSI values, predictions, and actual values for comparison
    print(f"Model: {name}")
    results[name] = list()
    
    for features, predicted, actual in zip(X_test.values, predictions, y_test.values):
        temp = list()
        temp.append([features] + [predicted] + [actual])
        results[name].append(temp)
        
        print(f"RSSI values: {features}")
        print(f"Predicted: {predicted}, Actual: {actual}")
        print('-' * 60)

```

#### Hyperparameter Tuning with Grid Search CV

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.multioutput import MultiOutputRegressor

def correct_params(params):
    return {key.replace('regressor__estimator__', ''): value for key, value in params.items()}

# Define a pipeline for each model to include in Grid Search
pipelines = {
    'Linear Regression': Pipeline([
        ('regressor', MultiOutputRegressor(LinearRegression()))
    ]),
    'Decision Tree': Pipeline([
        ('regressor', MultiOutputRegressor(DecisionTreeRegressor()))
    ]),
    'Random Forest': Pipeline([
        ('regressor', MultiOutputRegressor(RandomForestRegressor()))
    ]),
    'Gradient Boosting': Pipeline([
        ('regressor', MultiOutputRegressor(GradientBoostingRegressor()))
    ]),
    'K-Neighbors': Pipeline([
        ('regressor', MultiOutputRegressor(KNeighborsRegressor()))
    ])
}

# Define parameter grids for each model
param_grids = {
    'Linear Regression': { # Linear Regression usually doesn't need hyperparameter tuning
    },
    'Decision Tree': {
        'regressor__estimator__max_depth': [None, 10, 20, 30],
        'regressor__estimator__min_samples_split': [2, 10, 20]
    },
    'Random Forest': {
        'regressor__estimator__n_estimators': [50, 100, 200],
        'regressor__estimator__max_features': ['auto', 'sqrt', 'log2']
    },
    'Gradient Boosting': {
        'regressor__estimator__n_estimators': [100, 200, 300],
        'regressor__estimator__learning_rate': [0.01, 0.1, 0.2],
        'regressor__estimator__subsample': [0.9, 1.0]
    },
    'K-Neighbors': {
        'regressor__estimator__n_neighbors': [3, 5, 7],
        'regressor__estimator__weights': ['uniform', 'distance']
    }
}

tuned_models = {}

# Setting up GridSearchCV
for name, pipeline in pipelines.items():
    grid_search = GridSearchCV(pipeline, param_grids[name], cv=5, verbose=2, scoring='neg_mean_squared_error')
    grid_search.fit(X_train, y_train)
    print(f"Best parameters for {name}: {grid_search.best_params_}")
    print(f"Best score for {name}: {grid_search.best_score_}")
    tuned_models[name] = correct_params(grid_search.best_params_)

    # Optionally, evaluate on X_test and y_test or visualize results as needed

print(tuned_models)
```
