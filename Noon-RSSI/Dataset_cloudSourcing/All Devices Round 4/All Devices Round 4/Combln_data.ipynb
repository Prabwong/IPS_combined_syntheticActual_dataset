```python
import pandas as pd

# Load the CSV files
file_paths = [
    r"D:\Prab\IPS-Project\synthetic_data_point\Noon-RSSI\Dataset_cloudSourcing\All Devices Round 4\All Devices Round 4\device_1_20240607_113940.csv",
    r"D:\Prab\IPS-Project\synthetic_data_point\Noon-RSSI\Dataset_cloudSourcing\All Devices Round 4\All Devices Round 4\device_2_20240607_113940.csv",
    r"D:\Prab\IPS-Project\synthetic_data_point\Noon-RSSI\Dataset_cloudSourcing\All Devices Round 4\All Devices Round 4\device_3_20240607_113940.csv",
    r"D:\Prab\IPS-Project\synthetic_data_point\Noon-RSSI\Dataset_cloudSourcing\All Devices Round 4\All Devices Round 4\device_4_20240607_113938.csv",
    r"D:\Prab\IPS-Project\synthetic_data_point\Noon-RSSI\Dataset_cloudSourcing\All Devices Round 4\All Devices Round 4\device_5_20240607_113941.csv"
]

# Read each file into a dataframe and store in a list
dataframes = [pd.read_csv(file, skiprows=[0]) for file in file_paths]
dataframes = [df.rename(columns=lambda x: x.strip()) for df in dataframes]

# Read each file into a dataframe and print column names
for dataframe in dataframes:
    print(f"Columns in {dataframe}: {dataframe.columns.tolist()}")

# Display the first few rows of each dataframe to understand their structure
print([len(dataframes[i].columns) for i in range(len(dataframes))])

unique_columns = []

for i in range(len(dataframes)):
    for j in dataframes[i].columns:
        unique_columns.append(j)
        
print(len(set(unique_columns)))

# for i in range(len(dataframes)):
#     print(dataframes[i].shape)

print(unique_columns)
```

```python
# Identify all unique columns across the dataframes
all_columns = set()
for df in dataframes:
    all_columns.update(df.columns)

primary_cols = ['Date & Time', 'Latitude', 'Longitude', 'Floor']
sorted_columns = primary_cols + [col for col in sorted(list(all_columns)) if col not in primary_cols]

# Initialize an empty dataframe with all unique columns
unified_dataframe = pd.DataFrame(columns=sorted_columns)

# Merge each dataframe into the unified dataframe, filling missing entries with -100
for df in dataframes:
    # Ensure each dataframe has all the unified columns, filling missing ones with -100
    df = df.reindex(columns=sorted_columns, fill_value=-100)
    unified_dataframe = pd.concat([unified_dataframe, df], ignore_index=True)

# Display the shape and first few rows of the unified dataframe to verify
print(unified_dataframe.shape)
print(unified_dataframe[['Date & Time', 'Latitude', 'Longitude', 'Floor']].head())
print("------------------------------------------------")
print(unified_dataframe.columns)
```

```python
# Save the unified dataframe to a CSV file
unified_dataframe.to_csv(r"D:\Prab\IPS-Project\synthetic_data_point\Noon-RSSI\Dataset_cloudSourcing\All Devices Round 4\All Devices Round 4\Round4_combined_data.csv", index=False)
```
